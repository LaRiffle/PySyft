# Proba of satisfying |c_i| <= 2

```python
def proba(i):
    cum = 0
    for j in range(1, i-1): # 1 to i-2
        cum += 2**j
    r = (15*cum + 16)/16**(i-1)
    return r
```

### Challenge
$c_i = CW^i \oplus TruthTable(s, \alpha_i)$

Strategy:

Sample tables: $2^{\nu + 1}$ options and get $c_i$

n = nb of bits

Nb of hypothesis = $2^{n(\nu + 1)}$

$c_i$ can only have at max $2^{2(\nu + 1)}$ values

If uniformly random, the proba of output in n samples on in a subset of $2^{2(\nu + 1)}$ values is:

For $\nu = 1$, $p = \dfrac{15 \sum_{i=1}^{i-2} 2^i + 16}{16^{i-1}}$

For any \nu, $p = \dfrac{15 \sum_{i=1}^{i-2} 2^i + 16}{16^{i-1}}$


# Collision in CW

```python
nu = 10
m = (1 + 2 * nu)
n = 32
def proba_resistance(m, n):
    p = 1
    for k in range(n):
        p *= (1 - k*2**(-m))
    #print(n, p)
    return p
proba_resistance(m, n)

def cumsum(x):
    acc = 0
    for i in range(x+1):
        acc += i
    return acc
def approx_proba_resistance(m, n):
    p = 1 - n*(n-1)/2*2**(-m)
    #print(n, p)
    return p
print(proba_resistance(m, n))
print(approx_proba_resistance(m, n))
```

```python
n_max = 32
plt.plot(range(n_max), [proba_resistance(m, n) for n in range(n_max)])
plt.show()

plt.plot(range(n_max), [approx_proba_resistance(m, n) for n in range(n_max)])
plt.show()

plt.plot(range(n_max), [proba_resistance(m, n) - approx_proba_resistance(m, n) for n in range(n_max)])
plt.show()
```
